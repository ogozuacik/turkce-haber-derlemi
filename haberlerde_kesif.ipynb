```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('derlemler/filtrelenmis_temizlenmis_derlem.csv.gz')
```

```python
# grafik çizimi için yardımcı metot
def window_average(x,N):
    low_index = 0
    high_index = low_index + N
    w_avg = []
    while(high_index<len(x)):
        temp = sum(x[low_index:high_index])/N
        w_avg.append(temp)
        low_index = low_index + N
        high_index = high_index + N
    return w_avg
```

```python
#toplam ay sayısı (23 yıl)
n_bins = 276 

# bir kelimenin bütün haberlerde toplam geçiş sayısı ve 1 aylık periyotlarda haberlerde bulunma durumu için metot
def kelime_frekansı(word):
    word = word.lower()
    word_freq = data.news.apply(lambda text: 1 if word in text else 0)
    sum_word = word_freq.rolling(300).sum()
    sum_word = sum_word[300:]
    sum_word = window_average(sum_word, int(len(sum_word)/(n_bins+1)))
    return sum(word_freq), sum_word
```

### Örnek 1

Grafiklerde seçili kelimelerin zamana göre haberlerde ne kadar geçtiğini göstermektedir. (1997-2019)
Lejand'da parantez içerisindeki değer kelimenin toplam geçtiği haber sayısıdır.

```python
w1 = 'Erdoğan'
w2 = 'Kılıçdaroğlu'
w3 = 'Bahçeli'
w4 = 'Baykal'

s_w1, sum_w1 = kelime_frekansı(w1)
s_w2, sum_w2 = kelime_frekansı(w2)
s_w3, sum_w3 = kelime_frekansı(w3)
s_w4, sum_w4 = kelime_frekansı(w4)
```

```python
dt = pd.date_range(start='1/1/1997', end='1/1/2020', freq='M', closed='left')

f = plt.figure()
f.set_size_inches(12, 6)
plt.plot(dt, sum_w1[:276][::-1], label=(w1 + ' (' + str(s_w1) + ')'))
plt.plot(dt, sum_w2[:276][::-1], label=(w2 + ' (' + str(s_w2) + ')'))
plt.plot(dt, sum_w3[:276][::-1], label=(w3 + ' (' + str(s_w3) + ')'))
plt.plot(dt, sum_w4[:276][::-1], label=(w4 + ' (' + str(s_w4) + ')'))
plt.xlabel('Tarih', fontsize=14)
plt.ylabel('Haber Sayısı', fontsize=14)
plt.legend(loc='upper right')
```
